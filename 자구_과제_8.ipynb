##1.
chiaing과 open addressing이다. chaining은 구현이 단순하며 적재율에 거의 영향을 받지 않는다는 장점을 갖고 있고 open addressing은 포인터 없이 테이블 내의 빈 공간을 활용하므로 메모리 효율이 높다는 장점이 있다. 그러나 chaining은 포인터를 이용하여 포인터를 저장하기 위한 추가 메모리 공간이 필요하고 open addressing은 테이블이 가득차면 크기를 재조정해야 한다는 단점을 가지고 있다.
